```python
#operations supported during file handling 
1. opening a file 
2. reading or writing a file 
3. closing a file 
```

```python
#open() functions : polymorphic functions : 
1. if the file is exiting it will open the file for reading
2. if the file is not present and need to be created then it will create new file for writing 
```

```python
#types of operations perfomed : 
read  : r
write   : always first pointer    : w
append  : a
```

```python
#open('file path',access mode(r/w/a))
f=open('sample.txt',"w")  #opening a file for writing 
if f:
    print("file created")
f.close()
```

```python
#writing operations 
f=open('test.doc',"a")  #opening a file for writing 
if f:
    print("file created")
    
content=input("enter content to store")
f.write(content)
f.close()
```

```python
f=open('test.txt',"a")  #opening a file for writing 
if f:
    print("file created")
    
content=input("enter content to store")
f.write(content)

#reading a content of file 
f=open('test.txt',"r")  #opening a file for reading

r=f.read()
print(r)
f.close()
```

```python
f=open('test.txt',"r")  #opening a file for reading

r=f.read()
print(r)
con=r.split()
print(con)
print("the total words are ",len(con))
f.close()
```
